# Unnamed Column 제거

```python
KB_APT_price.columns.str.contains('^Unnamed')
```

```python
KB_APT_price.columns
```

# 시도지역 추출

```python
KB_APT_price = KB_APT_price[['구분', '전국', '서울','부산', '대구', '인천', '광주', '대전', \
                             '울산', '수도권', '세종', '경기','강원','충북', '충남', \
                             '전북', '전남', '경북', '경남', '제주/\n서귀포']]
KB_APT_price = KB_APT_price.rename(columns={'제주/\n서귀포': '제주'})
KB_APT_price
```

# 날짜 문자열을 Datetime type 으로 변환

```python
KB_APT_price["구분"]
```

```python
KB_APT_price["구분"].astype("str")
```

```python
KB_APT_price["구분"].astype("str").str.replace(".0$", "")
```

```python
year_str = ""

for date_string in KB_APT_price["구분"].astype("str").str.replace(".0$", ""):
    splitted_date_string = date_string.split('.')
    if len(splitted_date_string) == 2:
        year_str = splitted_date_string[0]
        print(date_string)
    else:
        print(year_str + "." + splitted_date_string[0])
```

```python
year_str = ""

date_string_list = []
for date_string in KB_APT_price["구분"].astype("str").str.replace(".0$", ""):
    splitted_date_string = date_string.split('.')
    if len(splitted_date_string) == 2:
        year_str = splitted_date_string[0]
        date_string_list.append(date_string)
    else:
        date_string_list.append(year_str + "." + splitted_date_string[0])
        
date_string_list
```

```python
year_str = ""

date_string_list = []
for date_string in KB_APT_price["구분"].astype("str").str.replace(".0$", ""):
    splitted_date_string = date_string.split('.')
    if len(splitted_date_string) == 2:
        year_str = splitted_date_string[0]
        
        if len(year_str) == 2:
            year_str = "19"+year_str #1986
        
        date_string_list.append(year_str + "." + splitted_date_string[1]) #1986.1
    else:
        date_string_list.append(year_str + "." + splitted_date_string[0]) #1986.2
        
date_string_list
```

```python
pd.to_datetime(date_string_list, format='%Y.%m')
```

```python
from pandas.tseries.offsets import MonthEnd

pd.to_datetime(date_string_list, format='%Y.%m') + MonthEnd(1)
```

```python
KB_APT_price.index = pd.to_datetime(date_string_list, format='%Y.%m') + MonthEnd(1)
KB_APT_price.index.name = "날짜"
KB_APT_price = KB_APT_price.drop(columns=["구분"])
KB_APT_price
```
