# Excel file 읽기

```python
# Sheet1 읽기
input_excel_file = "Rawdata/지역별_아파트_실거래가격지수.xlsx"
KAB_APT_tradeIndex2 = pd.read_excel(input_excel_file, sheet_name='Sheet1')
KAB_APT_tradeIndex2
```

```python
# 첫 5줄만 읽기
input_excel_file = "Rawdata/지역별_아파트_실거래가격지수.xlsx"
KAB_APT_tradeIndex2_info = pd.read_excel(input_excel_file, sheet_name='Sheet1', nrows=5)

# Unnamed 로 시작되는 column 제외하기
contains_bool = ~KAB_APT_tradeIndex2_info.columns.str.contains('^Unnamed')
KAB_APT_tradeIndex2_info = KAB_APT_tradeIndex2_info.loc[:, contains_bool]
KAB_APT_tradeIndex2_info
```

```python
# skiprows - 10줄 건너뛰고 읽기
input_excel_file = "Rawdata/지역별_아파트_실거래가격지수.xlsx"
KAB_APT_tradeIndex2 = pd.read_excel(input_excel_file, sheet_name='Sheet1', skiprows = range(0,10))
KAB_APT_tradeIndex2
```

# - (dash) 를 NaN 으로 변환

```python
KAB_APT_tradeIndex2 = KAB_APT_tradeIndex2.replace({'-': None})
KAB_APT_tradeIndex2
```

# Header 공백제거

```python
KAB_APT_tradeIndex2.columns
```

```python
KAB_APT_tradeIndex2.columns.str.replace(" ","")
```

```python
KAB_APT_tradeIndex2.columns = KAB_APT_tradeIndex2.columns.str.replace(" ","")
KAB_APT_tradeIndex2.head()
```

# Row index 설정

```python
KAB_APT_tradeIndex2 = KAB_APT_tradeIndex2.set_index(["지역"])
KAB_APT_tradeIndex2.head()
```

# 행열바꾸기 (Transpose)

```python
KAB_APT_tradeIndex2 = KAB_APT_tradeIndex2.T
KAB_APT_tradeIndex2.head()
```

# 날짜 문자열 Datetime 으로 type 변환

```python
KAB_APT_tradeIndex2.index
```

```python
pd.to_datetime(KAB_APT_tradeIndex2.index, format="%Y년%m월")
```

```python
from pandas.tseries.offsets import MonthEnd

KAB_APT_tradeIndex2.index = pd.to_datetime(KAB_APT_tradeIndex2.index, format="%Y년%m월") + MonthEnd(1)
KAB_APT_tradeIndex2.index.name = "날짜"
KAB_APT_tradeIndex2.index
```

```python
KAB_APT_tradeIndex2
```
